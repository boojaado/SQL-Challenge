```python
import os
import pandas as pd
import numpy as np
```

```python
os.listdir("data/")
```

```python
prefix = "data/"

files = os.listdir("data/")
dfs = []

for file in files:
    path = prefix + file
    df = pd.read_csv(path)
    dfs.append({"file": file, "df": df})
```

```python
for data in dfs:
    print(data["file"])
    display(data["df"].head())
    print(data["df"].info())
    print()
    print()
```
